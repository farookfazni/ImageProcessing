```python
import cv2
img = cv2.imread('F:/CST17020/python/ImageProcessingProject/my extra works/id/3.jpg')
img = cv2.resize(img,(1029,644))
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
bfilter = cv2.bilateralFilter(gray, 11, 17, 17) #Noise reduction
edged = cv2.Canny(bfilter, 30, 200) #Edge detection
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
import numpy as np


width,height= 500,100
#poins for id
pts1 = np.float32([[550,348],[880,348],[550,395],[880,395]])
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

#poins for name
pts3 = np.float32([[550,260],[950,260],[550,300],[950,300]])
pts4 = np.float32([[0,0],[width,0],[0,height],[width,height]])

#poins for faculty
pts5 = np.float32([[550,420],[980,420],[550,460],[980,460]])
pts6 = np.float32([[0,0],[width,0],[0,height],[width,height]])

#poins for degree
pts7 = np.float32([[550,500],[980,500],[550,620],[980,620]])
pts8 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix1=cv2.getPerspectiveTransform(pts1, pts2)
matrix2=cv2.getPerspectiveTransform(pts3, pts4)
matrix3=cv2.getPerspectiveTransform(pts5, pts6)
matrix4=cv2.getPerspectiveTransform(pts7, pts8)

imgId=cv2.warpPerspective(bfilter,matrix1,(width,height))
imgName=cv2.warpPerspective(bfilter,matrix2,(width,height))
imgFaculty=cv2.warpPerspective(bfilter,matrix3,(width,height))
imgDeg=cv2.warpPerspective(bfilter,matrix4,(width,height))

cv2.imshow("imgId",imgId)
cv2.imshow("imgName",imgName)
cv2.imshow("imgFaculty",imgFaculty)
cv2.imshow("imgDeg",imgDeg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
