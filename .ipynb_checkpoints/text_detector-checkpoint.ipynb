```python
import cv2
img = cv2.imread('F:/CST17020/python/ImageProcessingProject/my extra works/id/1.jpg')
img = cv2.resize(img,(1029,644))
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
import numpy as np
width,height= 950,580
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

pts1 = np.float32([[540,220],[950,220],[540,580],[950,580]])
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix1=cv2.getPerspectiveTransform(pts1, pts2)

imgId=cv2.warpPerspective(img,matrix1,(width,height))
gray1 = cv2.cvtColor(imgId,cv2.COLOR_BGR2GRAY)
```

```python
import pytesseract

pytesseract.pytesseract.tesseract_cmd = 'F:\\Program Files\\Tesseract-OCR\\tesseract.exe'

ret, thresh1 = cv2.threshold(gray1, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)
  

rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (18, 18))
  
# Appplying dilation on the threshold image
dilation = cv2.dilate(thresh1, rect_kernel, iterations = 1)
  
# Finding contours
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL, 
                                                 cv2.CHAIN_APPROX_NONE)
  
# Creating a copy of image
imI = imgId.copy()
im2 = img.copy()
  
# A text file is created and flushed
file = open("recognized.txt", "w+")
file.write("")
file.close()
  
# Looping through the identified contours
# Then rectangular part is cropped and passed on
# to pytesseract for extracting text from it
# Extracted text is then written into the text file
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
#     x, y, w, h =540,220,950,580
      
    # Drawing a rectangle on copied image
    rect = cv2.rectangle(imI, (x, y), (x + w, y + h), (0, 255, 0), 2)
      
    # Cropping the text block for giving input to OCR
    cropped = imI[y:y + h, x:x + w]
      
    # Open the file in append mode
    file = open("recognized.txt", "a")
      
    # Apply OCR on the cropped image
    text = pytesseract.image_to_string(cropped)
    print(text)
# print(text)
#     Appending the text into file
    file.write(text)
    file.write("\n")
      
#     Close the file
    file.close
cv2.imshow('image',imgId)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
